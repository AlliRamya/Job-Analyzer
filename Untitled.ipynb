```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

url = "https://remoteok.com/remote-data-jobs"

headers = {"User-Agent": "Mozilla/5.0"}
response = requests.get(url, headers=headers)

soup = BeautifulSoup(response.text, "html.parser")

# ✅ IMPORTANT LINE (you missed this)
job_rows = soup.find_all("tr", class_="job")

jobs = []

for job in job_rows:
    # 🚨 Skip placeholder rows
    if "placeholder" in job.get("class", []):
        continue

    title_tag = job.find("a", class_="preventLink") or job.find("h2")
    company_tag = job.find("h3")
    location_tag = job.find("div", class_="location")
    tags = job.find_all("span", class_="tag")

    if title_tag:
        jobs.append({
            "Title": title_tag.text.strip(),
            "Company": company_tag.text.strip() if company_tag else "Unknown",
            "Location": location_tag.text.strip() if location_tag else "Remote",
            "Skills": ", ".join([tag.text.strip() for tag in tags]) if tags else "Not Mentioned"
        })

print("Total rows found:", len(job_rows))
print("Collected jobs:", len(jobs))

df = pd.DataFrame(jobs)
print(df.head())
```

```python
import requests
import pandas as pd

url = "https://remoteok.com/api"

response = requests.get(url)
data = response.json()

jobs = []

for job in data[1:]:  # skip metadata
    jobs.append({
        "Title": job.get("position"),
        "Company": job.get("company"),
        "Location": job.get("location"),
        "Skills": ", ".join(job.get("tags", []))
    })

df = pd.DataFrame(jobs)
print(df.head())
```

```python
df.dropna(inplace=True)
df["Skills"] = df["Skills"].str.lower()
```

```python
from collections import Counter

all_skills = []

for skills in df["Skills"]:
    all_skills.extend(skills.split(", "))

skill_count = Counter(all_skills)

top_skills = skill_count.most_common(10)

print("Top Skills:")
for skill, count in top_skills:
    print(skill, count)
```

```python
location_counts = df["Location"].value_counts()

print("\nTop Locations:")
print(location_counts.head(10))
```

```python
df.to_csv("job_data.csv", index=False)
```

```python
import matplotlib.pyplot as plt

skills, counts = zip(*top_skills)

plt.bar(skills, counts)
plt.xticks(rotation=45)
plt.title("Top Skills in Demand")
plt.show()
```
